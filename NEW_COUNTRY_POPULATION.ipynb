# VISUALIZATION

POPULATION TREND IN A SPECIFIC COUNTRY (EG: AFGANISTAN)

```python
import matplotlib.pyplot as plt

country = 'Afghanistan'
country_data = data[data['Country Name'] == country]

country_data = country_data.drop('Country Name', axis=1).transpose()

country_data.index = country_data.index.astype(str)

plt.figure(figsize=(14, 7))
plt.plot(country_data.index, country_data.values, marker='o', label=country)

plt.title(f'Population Trend for {country}', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)

plt.xticks(ticks=range(len(country_data.index)), labels=country_data.index, rotation=90, fontsize=10)

plt.grid(True)
plt.legend()

plt.show()
```

PLOT FOR MULTIPLE COUNTRIES

```python
countries = ['United Arab Emirates', 'Argentina', 'India'] 

plt.figure(figsize=(12, 6))

for country in countries:
    country_data = data[data['Country Name'] == country].drop('Country Name', axis=1).transpose()
    plt.plot(country_data.index, country_data.values, marker='o', label=country)

plt.title('Population Trends for Multiple Countries')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

BAR PLOT OF POPULATION OF TOP 10 COUNTRY IN A SPECIFIC YEAR 

```python
year = '2020'
popltn = data[['Country Name', year]].sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(popltn['Country Name'], popltn[year], color='skyblue')
plt.title(f'Population for Top 10 Countries in {year}')
plt.xlabel('Population')
plt.ylabel('Country')
plt.gca().invert_yaxis()
plt.show()
```

POPULATION GROWTH RATE OVER TIME FOR A COUNTRY

```python
country = 'Bahrain'
country_data = data[data['Country Name'] == country].drop('Country Name', axis=1).transpose()

growth_rate = country_data.pct_change().fillna(0) * 100  

plt.figure(figsize=(12, 6)) 
plt.plot(growth_rate.index, growth_rate.values, marker='o', color='green', label=f'{country} Growth Rate')

plt.title(f'Population Growth Rate for {country}', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Growth Rate (%)', fontsize=12)

plt.xticks(ticks=range(0, len(growth_rate.index), 5), labels=growth_rate.index[::5], rotation=45, fontsize=10)

plt.grid(True)
plt.legend()

plt.show()

```

HEATMAP FOR POPUATION OF MULTIPLE COUNTRIES 

```python
import seaborn as sns

countries = ['Bhutan', 'Finland', 'Hungary', 'India', 'China', 'Brazil']

subset_data = data[data['Country Name'].isin(countries)].set_index('Country Name').transpose()

subset_data.index = pd.to_numeric(subset_data.index)

plt.figure(figsize=(10, 6))
sns.heatmap(subset_data, cmap='YlGnBu', annot=False, linewidths=0.5)
plt.title('Population Heatmap for Selected Countries')
plt.xlabel('Country')
plt.ylabel('Year')
plt.show()
```

LINE PLOT OF POPULATION GROWTH FOR MULTIPLE COUNTRIES

```python
countries = ['Indonesia', 'Italy', 'Japan', 'Maldives', 'Mexico']

plt.figure(figsize=(14, 7)) 

for country in countries:
    country_data = data[data['Country Name'] == country].drop('Country Name', axis=1).transpose()
    growth_rate = country_data.pct_change().fillna(0) * 100 
    plt.plot(growth_rate.index, growth_rate.values, label=country)

plt.title('Population Growth Rate Comparison for Selected Countries', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Growth Rate (%)', fontsize=12)

plt.xticks(ticks=range(0, len(growth_rate.index), 5), labels=growth_rate.index[::5], rotation=45, fontsize=10)

plt.grid(True)
plt.legend()

plt.show()


```

# PREDICTIVE ANALYTICS

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

country = input("Enter the country name: ") 
future_years = int(input("Enter the number of future years for prediction: ")) 

country_data = data[data['Country Name'] == country].drop('Country Name', axis=1).transpose()

x = np.array(country_data.index).astype(int).reshape(-1, 1) 
y = np.array(country_data.values).reshape(-1, 1)

split_year = -10
x_train, x_test = x[:split_year], x[split_year:]
y_train, y_test = y[:split_year], y[split_year:]

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print(f"\nPredicted vs Actual population for {country} (historical data from 2019 to 2022):")
for year, actual, predicted in zip(x_test, y_test, y_pred):
    if year[0] >= 2019 and year[0] <= 2022: 
        print(f"Year {year[0]}: Actual population = {actual[0]:.2f}, Predicted population = {predicted[0]:.2f}")

last_year = int(x[-1]) 
future_year_range = np.arange(last_year + 1, last_year + future_years + 1).reshape(-1, 1)
future_predictions = model.predict(future_year_range)

print(f"\nPredicted population for {country} (future years):")
for year, predicted in zip(future_year_range, future_predictions):
    print(f"Year {year[0]}: Predicted population = {predicted[0]:.2f}")

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)   

print("\n--- Accuracy Metrics ---")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"RÂ² Score: {r2:.2f}")
print(f"Accuracy: {r2 * 100:.2f}%")
```
